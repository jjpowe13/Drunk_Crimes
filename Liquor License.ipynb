```python
liqlic = pd.read_csv('D:\BU\Mod4\Analytics\Liquor_License_Data.csv')
liqlic = liqlic.drop(liqlic.columns[3:21], axis=1)
liqlic = liqlic.drop(liqlic.columns[4:9], axis=1)
liqlic = liqlic.reset_index(drop=True)

liqlic.head()
```
